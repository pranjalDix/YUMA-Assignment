```python
import pandas as pd      
import seaborn as sns
import matplotlib.pyplot as plt  
import numpy as np
```

```python
data = pd.read_csv('sales_data.csv')
data
```

```python
data.info()
```

```python
# Removing all NaN values from CustomerId

data = data.dropna(subset=['CustomerID', 'TransactionDate', 'PricePerUnit', 'TotalAmount'])
```

```python
data
```

```python
# Changing format of TransactionDate to DateTime

data['TransactionDate'] = pd.to_datetime(data['TransactionDate'])
```

```python
data.info()
```

```python
# Remapping correctly ProductID column to the particular ProductCategory name.

# dicitionary
category_to_id = {
    'Grocery': 2001,
    'Toys': 2002,
    'Home Decor': 2003,
    'Fashion': 2004,
    'Electronics': 2005
}

# Updating column
data['ProductID'] = data['ProductCategory'].map(category_to_id)
```

```python
data
```

```python
# Handling negative quantity values

data['Quantity'] = data['Quantity'].abs()
```

```python
data
```

```python
# Product being returned
```

```python
# Recalculating TotalAmount Column

data['TotalAmount'] = data['Quantity'] * data['PricePerUnit']
```

```python
data
```

```python
# Removing fields where quantity values is 0

data = data[data['Quantity'] != 0]
data
```

```python
# Updating Discount NaN values

data['DiscountApplied'] = data['DiscountApplied'].fillna(0)
```

```python
data
```

```python
# Updating Negative TrustPointsUsed values to positive

data['TrustPointsUsed'] = data['TrustPointsUsed'].abs()
```

```python
data
```

```python
# Updating PaymentMethod column

def update_payment_method(row):
    if pd.isna(row['PaymentMethod']):
        if row['TrustPointsUsed'] != 0:
            return 'Trust Points'
        else:
            return 'Credit Card/Cash'
    else:
        return row['PaymentMethod']

# Apply the function to each row
data['PaymentMethod'] = data.apply(update_payment_method, axis=1)
```

```python
data
```

```python
# Adding a new column which is total amount to be paid by the customer

data['AmountToBePaid'] = data['TotalAmount'] * (1 - data['DiscountApplied'] / 100)
```

```python
data
```

# Data Aggregation or Grouping

### 1. total number of Transactions

```python
transactions = data['TransactionID'].count()
transactions
```

### 2. total number of Unique Customers

```python
customers = data['CustomerID'].nunique()
customers
```

### 3. Total number of purchases by each customer

```python
purchases = data['CustomerID'].groupby(data['CustomerID']).count()
purchases
```

### 4. Total number of unique products

```python
products = data['ProductID'].nunique()
products
```

### 5. Sales by each product category

```python
sales_by_each_category = data.groupby('ProductCategory')['AmountToBePaid'].sum().sort_values(ascending = False)
sales_by_each_category
```

### 6. Total quantities of each product category

```python
quantity_product_category = data.groupby('ProductCategory')['Quantity'].sum().sort_values(ascending = False)
quantity_product_category
```

### 7. Total quantites sold

```python
total_quantity = data['Quantity'].sum()
total_quantity
```

### 8. Total Trust points used by each Customer

```python
trust_points_by_each_customer = data.groupby('CustomerID')['TrustPointsUsed'].sum().sort_values(ascending = False)
trust_points_by_each_customer
```

### 9. Total amount paid by each Customer

```python
amount_paid_by_each_customer = data.groupby('CustomerID')['AmountToBePaid'].sum().sort_values(ascending = False)
amount_paid_by_each_customer
```

### 10. Max, Min, Average Discount applied

```python
maxi_dis = data['DiscountApplied'].max()
maxi_dis
```

```python
mini_dis = data['DiscountApplied'].min()
mini_dis
```

```python
avg_dis = data['DiscountApplied'].mean()
avg_dis
```

### 11. Payment Method transactions Grouping

```python
payment_method_count = data.groupby('PaymentMethod')['TransactionID'].count().sort_values(ascending = False)
payment_method_count
```

### 12. Calculate the percentage for each payment type

```python
total_transactions = payment_method_count.sum()
payment_method_percentage = (payment_method_count / total_transactions) * 100

payment_method_percentage = payment_method_percentage.round(0).astype(int).astype(str) + '%'

payment_method_percentage
```

### 13. Number of transactions for each day

```python
transactions_per_day = data.groupby(data['TransactionDate'].dt.date)['TransactionID'].count()
transactions_per_day
```

### 14. Create Time frame groups and calculate the total sales

```python
bins = [0, 6, 12, 18, 24]
labels = ['00:00-06:00', '06:00-12:00', '12:00-18:00', '18:00-00:00']
data['Hour'] = data['TransactionDate'].dt.hour
data['TimeFrame'] = pd.cut(data['Hour'], bins=bins, labels=labels, right=False)
sales_per_time_frame = data.groupby('TimeFrame')['TotalAmount'].sum()
sales_per_time_frame
```

### 15. Recency, Frequency and Monetary Value

```python
current_date = pd.to_datetime('2024-08-12')

# Calculate Recency: days since the last purchase
rfm = data.groupby('CustomerID').agg({
    'TransactionDate': lambda x: (current_date - x.max()).days,
    'CustomerID': 'count',
    'TotalAmount': 'sum'
}).rename(columns={'TransactionDate': 'Recency', 
                   'CustomerID': 'Frequency', 
                   'TotalAmount': 'Monetary'})

rfm
```

# Linear Projection of Data

```python
from sklearn.linear_model import LinearRegression

X = data[['PricePerUnit']].values 
Y = data['TotalAmount'].values    

model = LinearRegression()

model.fit(X, Y)

Y_pred = model.predict(X)

Y_pred
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y, color='blue', label='Actual Data')

# Plotting regression line
plt.plot(X, Y_pred, color='red', linewidth=2, label='Fitted Line')

plt.xlabel('Price Per Unit')
plt.ylabel('Total Amount')
plt.title('Price Per Unit vs. Total Amount with Linear Regression Line')
plt.legend()

plt.show()
```
